```python
with open("data.txt","r") as file:
    print(file.readable())
    print(file.name)
    print(file.mode)
    print(file.closed)
    print(file.tell())
    file.seek(5)
    print(file.tell())
```

EXCEPTION HANDLING

```python
try:
    with open("data.txt", "r") as file:
        consent = file.read()
        print(consent)

except FileNotFoundError as e:
    print(e)
```

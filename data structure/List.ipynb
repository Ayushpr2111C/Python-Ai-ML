```python
apps=['google', 'youtube', 'twitter', 'linkedin', 'instagram']
print(apps[2])
print(apps[-3])
print(apps[1:5])
print(apps[1:8:2])
print(apps[:5])
print(apps[4:])
print(apps[:])
print(apps[-5:-1])
```

```python
from random import random, randrange, randint, shuffle, choice
```

```python
print("numbers between 0 and 1:", random())
print("number between 1 and 10:", randrange(1, 11))
print("random choice from list", choice([1, 2, 3, 4, 5]))

name = ["john", "Jane", "Jack", "Jill"]
shuffle(name)
print("shuffled name:", name)
```

```python
from datetime import datetime as dt
```

```python
print("Current date and time:", dt.now())
print("Current year:", dt.now().year)
print("Current month:", dt.now().month)
print("Current day:", dt.now().day)
print("Current hour:", dt.now().hour)
print("Current minute:", dt.now().minute)
print("Current second:", dt.now().second)
print("Current microsecond:", dt.now().microsecond)
print("Current timezone:", dt.now().tzinfo)
```

```python
print("Current date", dt.now().strftime("%B %d %Y"))
print("Current time", dt.now().strftime("%I:%M:%S %p"))
print("current month", dt.now().strftime("%B"))
print("current month", dt.now().strftime("%b"))
print("current day", dt.now().strftime("%A"))
print("current day", dt.now().strftime("%a"))
print("current year", dt.now().strftime("%Y"))
```

```python
import os
```

```python
print("Current working directory:", os.getcwd())
print("List of files in current directory:", os.listdir())
```

```python
total_files=0
total_folders=0
for path, files, folders in os.walk(os.getcwd()):
    print("Current path:", path)
    print("Files in current path:", files)
    print("Folders in current path:", folders)
    total_files += len(files)
    total_folders += len(folders)
print("Total files in current directory:", total_files)
print("Total folders in current directory:", total_folders)
```

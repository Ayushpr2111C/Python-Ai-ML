```python
import math
```

```python
print(math.sqrt(16))
print(math.pi)
print(math.e)
print(math.factorial(5))
print(math.gcd(12, 16))
print(math.lcm(12, 16))
print(math.isqrt(20))
```

```python
from random import random, randrange, randint, shuffle, choice
```

```python
print("numbers between 0 and 1:", random())
print("number between 1 and 10:", randrange(1, 11))
print("random choice from list", choice([1, 2, 3, 4, 5]))

name = ["john", "Jane", "Jack", "Jill"]
shuffle(name)
print("shuffled name:", name)
```

```python
from datetime import datetime as dt
```

```python
print("Current date and time:", dt.now())
print("Current year:", dt.now().year)
print("Current month:", dt.now().month)
print("Current day:", dt.now().day)
print("Current hour:", dt.now().hour)
print("Current minute:", dt.now().minute)
print("Current second:", dt.now().second)
print("Current microsecond:", dt.now().microsecond)
print("Current timezone:", dt.now().tzinfo)
```

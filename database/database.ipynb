```python
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
engine = create_engine('sqlite:///products.db')
```

```python
Base = declarative_base()
```

```python
class Product(Base):
    __tablename__ = "product"

    id = Column(Integer, primary_key=True)
    name = Column(String, nullable=False)
    price = Column(Float, nullable=False)
    category = Column(String, nullable=False)
```

```python
Base.metadata.create_all(engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
products = [
    Product(name="Laptop", price=45000.0, category="Electronics"),
    Product(name="Mouse", price=800.0, category="Electronics"),
    Product(name="Phone", price=25000.0, category="Electronics"),
    Product(name="Keyboard", price=1500.0, category="Electronics"),
    Product(name="Headphones", price=3000.0, category="Electronics")
]

session.add_all(products)
session.commit()

print("5 products added successfully")
```

```python
all_products = session.query(Product).all()

for p in all_products:
    print(f"ID: {p.id}, Name: {p.name}, Price: {p.price}, Category: {p.category}")
```

```python
#read one product
product = session.query(Product).filter_by(name="Phone").first()

if product:
    print("Selected product:", product.name, product.price, product.category)
else:
    print("Product not found")
```

```python
#update
product = session.query(Product).filter_by(name="Phone").first()

if product:
    product.price = 26000.0
    product.category = "Mobile"
    session.commit()

updated_product = session.query(Product).filter_by(name="Phone").first()
print("Updated product:", updated_product.name, updated_product.price, updated_product.category)
```

```python
updated_product = session.query(Product).filter_by(name="Phone").first()
if updated_product:
    print(updated_product.name, updated_product.price, updated_product.category)
else:
    print("Product not found")
```

```python
product_to_delete = session.query(Product).filter_by(name="Mouse").first()
if product_to_delete:
    session.delete(product_to_delete)
    session.commit()
remaining_products = session.query(Product).all()

print("After delete:")
for p in remaining_products:
    print(f"ID: {p.id}, Name: {p.name}, Price: {p.price}, Category: {p.category}")
```

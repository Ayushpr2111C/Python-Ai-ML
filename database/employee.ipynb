```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import sessionmaker, declarative_base
```

```python
engine = create_engine('sqlite:///employee.db')
Base = declarative_base()
```

```python
class Employee(Base):
    __tablename__ = 'employee'
    id = Column(Integer, primary_key=True)
    name = Column(String)
    age = Column(Integer)
    gender = Column(String)
    post = Column(String)

Base.metadata.create_all(engine)
session = sessionmaker(bind=engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
employees = [
    Employee(id= 4, name="Michael", age=30, gender="Male", post="King"),
    Employee(id= 5, name="Stephe", age=25, gender="Male", post="Best"),
    Employee(id= 6, name="Andres", age=39, gender="Male", post="GOD")
]
session.add_all(employees)
session.commit()
```

```python
all_products = session.query(Employee).all()
for p in all_products: 
    print(p.id, p.name, p.age, p.gender, p.post)
```

```python
product = session.query(Employee).filter_by(id=2).first()

product.name = "John Doe"
product.age = 30
product.gender = "Male"
product.post = "Manager"

session.commit()
```

```python
num = int(input("Enter a number: "))
is_prime = True
if num < 2:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

SOME BEUTIFUL CODES

```python
num1=int(input("Enter first number: "))
orig = num1
rev = 0

while num1 > 0:
    digit = num1 % 10
    rev = rev * 10 + digit
    num1 //= 10

print(f"Original number: {orig}")
print(f"Reversed number: {rev}")
```

```python
# check a number is palindrome or not
num1=int(input("Enter first number: "))
orig = num1
rev = 0
 
while num1 > 0:
    digit = num1 % 10
    rev = rev * 10 + digit
    num1 //= 10 

if orig == rev:
    print(f"{orig} is a palindrome number.")
else:
    print(f"{orig} is not a palindrome number.")
    
```

CHECK WEATHER AN NO. IS AMSTRONG OR NOT

AMSTRONG NO. IS LIKE 153 has 3 no.'s so the power should be 3 AND IF 1^3 + 5^3 + 3^3 = 153  THEN THE NO. IS AMSTRING

```python
num=int(input("Enter a number: "))
orig = num
digits = len(str(num))

sum_of_powers = 0

while num > 0:
    digit = num % 10
    sum_of_powers += digit ** digits
    num //= 10

if sum_of_powers == orig:
    print(f"{orig} is an Armstrong number.")
else:
    print(f"{orig} is not an Armstrong number.")
```

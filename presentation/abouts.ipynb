```python
df = pd.read_csv('sales_data.csv')
df.to_csv('sales_data.csv', index=False)
df.head()
```

```python

sales= df.groupby("State")["Sales"].sum().reset_index()
print(sales)
from numpy import average


sales= df.groupby("State")["Sales"].sum()
top_state=sales.idxmax()
top_state_sales = sales.max()
average_sales = sales.mean()
top_ = ((top_state_sales - average_sales)/average_sales)*100

print(f"Top State: {top_state}")
print(f"Top State Sales: ₹{top_state_sales:,.2f}")
print(f"Average Sales: ₹{average_sales:,.2f}")
print(f"{top_state} exceeds the average by {top_:.1f}%")
```

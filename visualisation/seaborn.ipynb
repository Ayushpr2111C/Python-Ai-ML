```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('Tips')
df
```

```python
sns.histplot(df, x='tip', kde=True, bins=30, color='red')
plt.title('Histogram of Tip')
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=df, x='day', bins=30, kde=True)
plt.title('Histogram of Day')
plt.xlabel('Day of the Week')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex',style='time')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.show()
```

```python
sns.lineplot(df, x='size', y='tip',marker='o')
plt.title('Line Plot of Tip vs Size')
plt.xlabel('Size')
plt.ylabel('Tip')
plt.show()
```

- genderwise count of per day
- bar plot of average bill by day

```python
sns.countplot(data=df, x="day", hue="sex")
plt.title('Count Plot of Sex vs Size')
plt.xlabel('Sex')
plt.ylabel('Size')
plt.show()
```

```python
sns.barplot(data=df, x='day', y='total_bill')
plt.title('Bar Plot of Total Bill vs Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.violinplot(data=df, x='day', y='tip', hue='sex', split=True, inner='quartile')
plt.title('Violin Plot of Tip vs Day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
sns.heatmap(data=df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(data=df, x='time')
plt.title('Count Plot of Size vs Time')
plt.xlabel('size')
plt.ylabel('Time')
plt.show()
```

```python
sns.barplot(data=df, x='smoker',y='tip')
plt.title('Smoker vs non-smoker')
plt.show()
```

```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df.head()
```

```python
fig = px.bar(df, x='abbrev', y='total',color='total', color_continuous_scale='Viridis',template='plotly_dark',
                 title='Total Car Crashes by State',
                 labels={'abbrev': 'State Abbreviation', 'total': 'Total Crashes'})
fig.show()
```

```python
top10 = df.sort_values('ins_premium', ascending=False).head(10)
fig = px.bar(top10, x='abbrev', y='ins_premium', 
                 color='ins_premium', 
                 color_continuous_scale='Viridis',
                 template='plotly_dark',
                 title='Top 10 States by Insurance Premium'
)
fig.show()
```

```python
fig = px.scatter(df, x='alcohol', y='speeding',
                color='total',
                size='total',
                template='plotly_dark',
                title='Alcohol vs. Speeding')
fig.show() 
```

```python
# line chart of statewise accidents
fig = px.line(df, x= 'abbrev', y='total', template='plotly_dark', 
              title='Statewise Accidents', markers='o')
fig.show()
```

```python
fig = px.pie(df, values='speeding', names='total', 
            template='plotly_dark',
            color_discrete_sequence=px.colors.sequential.Viridis, 
            title='Speeding Accidents by State')

fig.update_traces(textposition='inside')

fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
plt.show()  # Display the figure.
```

```python
df = sns.load_dataset('iris')
df
```

```python
#histogram - distribution of a single variable
plt.hist(df['sepal_length'], color='skyblue', edgecolor='black')
plt.title('distribution of sepal length')
plt.xlabel('sepal length')
plt.ylabel('count')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df['sepal_length'], df['petal_length'],  color='green')
plt.show()
```

```python
plt.bar( df['petal_length'][:50], color='red')
plt.title('petal length of first 50 flowers')
plt.xlabel('index')
plt.ylabel('petal length')
plt.show()
```

```python
plt.bar(df['sepal_length'][:50], df['petal_length'][:50], color='red')
plt.title('relationship between sepal length and petal length')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.show()
```

```python
mean_value= df.groupby('species').mean()
```

```python
mean_value.plot(kind='bar', figsize=(5, 5))
plt.title('mean value of each species')
plt.xlabel('species')
plt.ylabel('mean value')
plt.show()
```
